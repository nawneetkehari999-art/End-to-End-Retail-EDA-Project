```python
import pandas as pd

df = pd.read_csv("../Superstore_Management_System2.csv")
df.head()
```

```python
# Clean column names (remove extra spaces)
df.columns = df.columns.str.strip()

# Rename incorrect column names
df.rename(columns={
    "Custommer Name": "Customer Name",
    "sale Amount": "Sales Amount",
    "profit": "Profit",
    "Supplier Name ": "Supplier Name"
}, inplace=True)

df.columns
```

### Fix Category Typos

```python
data1 = df["Category"] = df["Category"].replace({
    "Office Supplise": "Office Supplies",
    "Electroincs": "Electronics"
})

data2 = df["Customer Segment"] = df["Customer Segment"].replace({
    "Corprate": "Corporate"
})
print(data1)
print(data2)
```
