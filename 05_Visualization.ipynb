```python
# importing important Libraries

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import datetime 
```

## UNIVARIATE ANALYSIS

```python
import pandas as pd

df = pd.read_csv("../Superstore_Management_System2.csv")
# df = pd.read_excel("your_file.xlsx")

import matplotlib.pyplot as plt # for crweating visualizations
```

```python
# Univariate Analysis is the simplest form of statistical analysis. As the name suggeste- "unit"
# means "one", and "variate" it is the analysis of a single variable.
print("\n=== CATEGORICAL VARIABLES ===\n")

categorical_cols = ['Customer Segment', 'Category', 'Region', 'Payment Mode', 'Delivery Status', 'Auto Reorder']

for col in categorical_cols:
    print(f"\n--- {col} ---")
    print(df[col].value_counts())

    plt.figure(figsize=(6, 6))
    df[col].value_counts().plot(kind='bar', color='purple', edgecolor='black', alpha=0.7)
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=30)
    plt.tight_layout()
    plt.show()
```

## Numerical  Variables Analysis

```python
print("\n=== NUMERICAL VARIABLES === \n")

numerical_cols = ['Quantity', 'Unit Price', 'Discount(%)', 'Sales Amount', 'Cost Price', 'Profit', 'Stock Left', 'Reorder Quantity' ]
for col in numerical_cols:
    print(f"\n --- {col} ---\n")
    print(f"Mean: {df[col].mean():.2f}")
    print(f"Median: {df[col].median():.2f}")
    print(f"Std Dev: {df[col].std():.2f}")
    print(f"Min: {df[col].min():.2f}")
    print(f"Max: {df[col].max():.2f}")


    # Visualization
    plt.figure(figsize=(8, 4))
    plt.subplot(1, 2, 1)
    plt.hist(df[col], bins=30, color='lightgreen', edgecolor='black', alpha = 0.7)
    plt.title(f'Distribution of {col}')

    plt.subplot(1, 2, 2)
    plt.boxplot(df[col])
    plt.title(f'Boxplot of {col}')

    plt.show()
```

## BIVARIATE ANALYSIS

```python
# Bivariate anlaysis is the statistical analysis of two variables at the same time. 
# The core purpose is to investigate the relationship between them.

title = "BIVARIATE ANALYSIS"
print("\n" + "+"*(len(title)+20))
print(" " *10 + title)
print("+"*(len(title)+20) + "\n")


category_performance = df.groupby('Category').agg({
    'Sales Amount': 'sum',
    'Profit': 'sum',
    'OrderId': 'count'
}).rename(columns={'OrderId': 'Order Count'})

category_performance['Profit Margin (%)'] = (
    category_performance['Profit']/ category_performance['Sales Amount']
)*100

print(category_performance,'\n\n')

print("\n === ====================================================================== ===\n")


# Visualization 
fig , axes = plt.subplots(2, 2, figsize =(15, 12))

# Sales By Category
category_performance['Sales Amount'].plot(kind='bar', ax=axes[0,0],color='lightblue', alpha=0.7, edgecolor='black')
axes[0,0].set_title("Total Sales by Category")
axes[0,0].tick_params(axis='x', rotation=0)

# Profit By Category
category_performance['Profit'].plot(kind='bar', ax=axes[0,1],color='lightcoral', alpha=0.7, edgecolor='black')
axes[0,1].set_title("Total Profit by Category")
axes[0,1].tick_params(axis='x', rotation=0)

# Profit Margin
category_performance['Profit Margin (%)'].plot(kind='bar', ax=axes[1,0],color='orange', alpha=0.7, edgecolor='black')
axes[1,0].set_title("Profit Margin by Category (%)")
axes[1,0].tick_params(axis='x', rotation=0)

# Order Count By Category
category_performance['Order Count'].plot(kind='bar', ax=axes[1,1],color='lightgreen', alpha=0.7, edgecolor='black')
axes[1,1].set_title("Total Order Count by Category")
axes[1,1].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()



```

## Regional Performance

```python
title ="REGIONAL PERFORMANCE"
print("\n" + "+"*(len(title)+20))
print(" " *10 + title)
print("+"*(len(title)+20) + "\n")

regional_performance = df.groupby('Region').agg({
    'Sales Amount': 'sum',
    'Profit': 'sum',
    'OrderId': 'count'
}).rename(columns={'OrderId': 'Order Count'})

# Profit Margin
regional_performance['Profit Margin (%)'] = (
    regional_performance['Profit'] /
    regional_performance['Sales Amount']
) * 100

print(regional_performance, '\n\n')

# Visualization
plt.figure(figsize=(12, 8))

# Sales Pie
plt.subplot(2, 2, 1)
regional_performance['Sales Amount'].plot(kind='pie', autopct='%1.1f%%',colors=['lightblue', 'lightgreen', 'lightcoral', 'lightsalmon'])
plt.title("Sales Distribution by Region")

# Profit Pie
plt.subplot(2, 2, 2)
regional_performance['Profit'].plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral', 'lightsalmon'])
plt.title("Profit Distribution by Region")

# Order Count
plt.subplot(2, 2, 3)
sns.barplot(x=regional_performance.index,
            y='Order Count',
            data=regional_performance, palette=['lightblue', 'lightgreen', 'lightcoral', 'lightsalmon'])
plt.title("Total Order Count by Region")
plt.xticks(rotation=30)

# Profit Margin
plt.subplot(2, 2, 4)
sns.barplot(x=regional_performance.index,
            y='Profit Margin (%)',
            data=regional_performance, palette=['lightblue', 'lightgreen', 'lightcoral', 'lightsalmon'])
plt.title("Profit Margin by Region (%)")
plt.xticks(rotation=30)

plt.tight_layout()
plt.show()
```

### Regional Performance

```python
title ="REGIONAL PERFORMANCE"
print("\n" + "+"*(len(title)+20))
print(" " *10 + title)
print("+"*(len(title)+20) + "\n")

regional_performance = df.groupby('Region').agg({
    'Sales Amount': 'sum',
    'Profit': 'sum',
    'OrderId': 'count'
}).rename(columns={'OrderId': 'Order Count'})

# Profit Margin
regional_performance['Profit Margin (%)'] = (
    regional_performance['Profit'] /
    regional_performance['Sales Amount']
) * 100

print(regional_performance, '\n\n')

# Visualization
plt.figure(figsize=(12, 8))

# Sales Pie
plt.subplot(2, 2, 1)
regional_performance['Sales Amount'].plot(kind='pie', autopct='%1.1f%%')
plt.title("Sales Distribution by Region")

# Profit Pie
plt.subplot(2, 2, 2)
regional_performance['Profit'].plot(kind='pie', autopct='%1.1f%%')
plt.title("Profit Distribution by Region")

# Order Count
plt.subplot(2, 2, 3)
sns.barplot(x=regional_performance.index,
            y='Order Count',
            data=regional_performance)
plt.title("Total Order Count by Region")
plt.xticks(rotation=30)

# Profit Margin
plt.subplot(2, 2, 4)
sns.barplot(x=regional_performance.index,
            y='Profit Margin (%)',
            data=regional_performance)
plt.title("Profit Margin by Region (%)")
plt.xticks(rotation=30)

plt.tight_layout()
plt.show()
```

### TIME SERIES ANALYSIS

```python
title = "TIME SERIES ANALYSIS"
print("\n" + "+"*(len(title)+20))
print(" " *10 + title)
print("+"*(len(title)+20)+ "\n")

# Ensure datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Extract Month & Year
# df['Order Date'] = pd.to_datetime(df['Order Date'])

# df['Order Month'] = df['Order Date'].dt.to_period('M')
df['Order Month'] = df['Order Date'].dt.to_period('M').dt.to_timestamp()


# Monthly trends
monthly_trend = df.groupby('Order Month').agg({
    'Sales Amount': 'sum',
    'Profit': 'sum',
    'OrderId': 'count'
}).rename(columns={'OrderId': 'Order Count'})

monthly_trend = monthly_trend.sort_index()

print("Monthly Trends:\n")
print(monthly_trend.tail(10), '\n\n')

# Visualization
plt.figure(figsize=(15, 10))

# Sales
plt.subplot(3, 1, 1)
monthly_trend['Sales Amount'].plot(kind='line', color='blue', marker='o')
plt.title('Monthly Sales Trend')
plt.ylabel('Sales Amount')
plt.grid(True)

# Profit
plt.subplot(3, 1, 2)
monthly_trend['Profit'].plot(kind='line', color='green', marker='o')
plt.title('Monthly Profit Trend')
plt.ylabel('Profit')
plt.grid(True)

# Order Count
plt.subplot(3, 1, 3)
monthly_trend['Order Count'].plot(kind='line', color='red', marker='o')
plt.title('Monthly Order Count Trend')
plt.ylabel('Order Count')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## ADVANCED ANALYSIS

```python

title = "CORRELATION ANALYSIS"
print("\n" + "+"*(len(title)+20))
print(" " *10 + title)
print("+"*(len(title)+20)+ "\n")

# Select numerical Columns for correlation 
numerical_df= df[['Quantity', 'Unit Price', 'Discount(%)', 'Sales Amount','Cost Price', 'Profit' ]]

# Correlation Matrix 
correlation_matrix = numerical_df.corr()
print("Correlation Matrix:")
print(correlation_matrix)

print("Correlation Matrix: \n \n")

# Heatmop 
plt.figure(figsize =(10, 8))
sns.heatmap(correlation_matrix, annot = True, cmap = 'coolwarm', center =0,
             square= True, linewidths =0.5)

plt.title("Correlation Matrix Heatmap")
plt.show()
plt.tight_layout()

```

### Customer Segment Analysis

```python
print("\n ================CORRELATION ANALYSIS ================\n") 



segment_analysis  = df.groupby('Customer Segment').agg({
    'Sales Amount': ['sum', 'mean'],
    'Profit': ['sum', 'mean'],
    'OrderId': 'count',
    'Discount(%)': 'mean'
}).round(2)

segment_analysis   .columns = ['Total Sales', 'Avg Sales', 'Total Profit', 'Avg Profit', 'Order Count', 'Avg Discount']
print(segment_analysis, '\n\n')

# Visualization
fig, axes = plt.subplots (2,2, figsize = (15, 10))

segment_analysis['Total Sales'].plot(kind= 'bar',ax = axes [0,0], color= 'lightblue')
axes[0,0].set_title ("Total Sales by Customer Segment")
axes[0,0].tick_params (axis = 'x', rotation = 30)


segment_analysis['Total Profit'].plot(kind= 'bar',ax = axes [0,1], color= 'lightgreen')
axes[0,1].set_title ("Total Profit by Customer Segment")
axes[0,1].tick_params (axis = 'x', rotation = 30)

segment_analysis['Order Count'].plot(kind= 'bar',ax = axes [1,0], color= 'orange')
axes[1,0].set_title ("Order Count by Customer Segment")
axes[1,0].tick_params (axis = 'x', rotation = 30)

segment_analysis['Avg Discount'].plot(kind= 'bar',ax = axes [1,1], color= 'red')
axes[1,1].set_title ("Average Discount by Customer Segment")
axes[1,1].tick_params (axis = 'x', rotation = 30)

plt.tight_layout()
plt.show()




```

### KEY INSIGHTS & SUMMARY

```python
print("\n ================ KEY INSIGHTS SUMMARY ================")

# Oberall Metrics
total_sales = df['Sales Amount'].sum()
total_profit = df['Profit'].sum()
totl_orders = df['OrderId'].nunique()
avg_profit_margin = (total_profit / total_sales)* 100

print(f"OVERALL PERFORMANCE: \n")
print(f"Total Sales: ${total_sales:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
print(f"Total Orders: {totl_orders}")
print(f"Average Profit Margin: {avg_profit_margin:.2f}%")


# Top Performing Categories

top_categories = df.groupby('Category')['Profit'].sum().sort_values(ascending =False)
print(f"\n TOP PERFORMING CATEGORIES: \n")
for i, (category, profit) in enumerate(top_categories.items(), 1):
    print(f"{i}. {category}: ${profit:,.2f}")

# Regional Performance
best_region = df.groupby('Region') ['Profit'].sum().idxmax()
print(f"TOP PERFORMING REGION: {best_region}")


# Delivery Prfirmance
delivery_stats = df['Delivery Status'].value_counts(normalize =True)*100
print(f"\n DELIVERY PERFORMANCE: \n")
for status, percentage in delivery_stats.items():
    print(f"{status}: {percentage:.1f}%")


# Customer Segment Analysis
best_segment = df.groupby('Customer Segment')['Profit'].sum().idxmax()
print(f"\n MOST PROFITABLECUSTOMER SEGMENT: {best_segment}")


```

### EXPORT RESULTS

```python
analysis_results ={
    'total_sales': total_sales,
    'total_profit': total_profit,
    'total_orders': totl_orders,
    'avg_profit_margin': avg_profit_margin,
    'top_categories': top_categories.index[0],
    'best_region': best_region,
    'best_customer_segment': best_segment
    }

# Convert to DataFrame for better visualization
results_df =pd.DataFrame([analysis_results])
results_df.to_csv("EDA _Analysis_Summary.CSV", index = False)
print("\n Analysis Summary saved to 'EDA_Analsis.CSV'")
```
